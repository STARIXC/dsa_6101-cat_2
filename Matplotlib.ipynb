```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

functional method

```python

#xlabel
plt.xlabel('X-assis')
plt.ylabel('Y-assis')
plt.title("X and Y axes")
plt.plot(x,y)
plt.show()
```

```python
# Multiplot on same canvas(number rows, col,plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
#object-oriented method

fig=plt.figure()
# left aix, botom, with and height
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X-assis')
axes.set_ylabel('Y-assis')
axes.set_title('X and Y axes')
axes.plot(x,y,'r')
```
